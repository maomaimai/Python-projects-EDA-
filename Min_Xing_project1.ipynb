Q.12 Show one instresting finding of your own which is not covered above. Show the code and write your finding as comment.

As an international student, I would like to check the approval rate on PERM applications for people who entered the US with F-1 and  H1-B visa.

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

data1 = data[(data['class_of_admission']=='H-1B') | (data['class_of_admission']=='F-1')][['class_of_admission','decision_date','case_status']]
data1['decision_date']=pd.to_datetime(data1['decision_date'],errors='coerce')
data1['decision_year']=data1['decision_date'].dt.year
data1.drop(['decision_date'],axis=1,inplace=True)
data2=data1.groupby(['decision_year','class_of_admission','case_status']).case_status.agg('count').to_frame('casecount').reset_index()
data3=data2.groupby(['decision_year','class_of_admission']).casecount.agg('sum').to_frame('totalcasenumber').reset_index()
data4=pd.merge(data2,data3,on=['decision_year','class_of_admission'],how='inner')
data4=data4[data4['case_status']!='Denied'].groupby(['decision_year','class_of_admission']).casecount.agg('sum').to_frame('approvecase').reset_index()
data5=pd.merge(data3,data4,on=['decision_year','class_of_admission'],how='inner')
data5['approval_rate']=data5['approvecase']/data5['totalcasenumber']
approval_rate_F1=data5['approval_rate'][::2]
approval_rate_H1B=data5['approval_rate'][1::2]
year=data5['decision_year'][::2]
plt.plot(year,approval_rate_F1,label='F-1')
plt.plot(year,approval_rate_H1B,label='H-1B')
plt.xlabel('year')
plt.ylabel('Approval Rate')
plt.legend()
plt.show()
```

From the above plot, we know for both types of visas, the change patterns of approval rate are similar: the approval rates were
declining from 2011 to 2013(Maybe this wass the lagged effect of the economic recession in 2008-2010). There was a sharp increase in approval rate in year 2014. After that, the approval rates were increasing at a very slow pace.
For the reason of why there a sharp increase in 2013, my guess is that 2013 was the year when President Obama inaugurated for his second term, and the labor market began to turn upwards.
